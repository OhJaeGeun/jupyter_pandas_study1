```python
a = '''
    pd.merge(left, right, on = "기준column", how = "left")
    left와 right는 변합할 두 dataframe을 대입함.
    on에는 병합 기준이 되는 column을 넣어줌.
    how에는 "left, right, inner, outer" <- 4가지 병합 방식 중 한가지를 선택함.
'''
```

```python
df_right = df2.drop([1, 3, 5, 7, 9], axis = 0)
df_right
```

```python
df_right = df_right.reset_index(drop = True)
```

```python
df_right
```

```python
df
```

```python
pd.concat([df, df_right], axis = 1)
```

```python
pd.merge(df, df_right, on = '이름', how = 'left')
```

```python
#pd.merge(df, df_right, on = '이름', how = 'right') 아래 코드와 동일
pd.merge(df_right, df, on = '이름', how = 'left')
```

```python
# inner 방식은 Dataframe에 모두 키 값이 존재하는 경우에만 병합
pd.merge(df, df_right, on = "이름", how = "inner")
```

```python
# outer 방식은 하나의 Dataframe에 모두 키 값이 존재하는 경우에만 병합
# outer 방식에서는 없는 값은 NaN으로 대입되어 짐.
pd.merge(df, df_right, on = '이름', how = 'outer')
```

```python
df_right.columns = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
df_right
```

```python
#pd.merge(df, df_right, on = '이름', how = 'outer') 불가
# df와 df_right를 병합하려면 df에는 '이름', df_right에는 'A'로 표기되어 있어서
# 기준을 left_on, right_on을 사용하여 병합할 수 있음
pd.merge(df, df_right, left_on = '이름', right_on = 'A', how = 'outer')
```

```python
def male_and_female(x):
    if x == '남자':
        return 1
    elif x == '여자':
        return 0
```

```python
#df['성별']
df['성별'].apply(male_and_female) # male_and_female : 콜백 함수
```

```python
# cm당 브랜드평판지수 구하기 : -> (브랜드 평판 지수 / 키)
```

```python
def cm_per_brand(df):
    value = df['브랜드평판지수'] / df['키']
    return value
```

```python
df['비율'] = df.apply(cm_per_brand, axis = 1)
df
```

```python
# lambda 함수
# - lambda는 한 줄로 작성하는 간단한 이름 없는 함수(식)
# - return을 별도로 명시하지 않음.
```

```python
male_and_female = lambda x : 1 if x == '남자' else 0
```

```python
# 25. 원핫인코딩(one-hot-encoding)
# - 원핫인코딩은 한개의 요소는 True,
# - 나머지 요소는 False로 만들어주는 기법
```

```python
blod_map = {'A' : 0, 'B' : 1, 'AB' : 2, 'O' : 3}
df['혈액형_code'] = df['혈액형'].map(blod_map)
```

```python
df
```

```python
df['혈액형_code'].value_counts()
```

```python
a = """
    - df['혈액형_code']를 머신러닝 알고리즘에 넣어 데이터를 예측하라고
      지시한다면 컴퓨터는 값들간의 관계를 스스로 형성하게 됨.
    - 만약 B형은 1, AB형은 2라는 값을 가지고 있을 때,
      컴퓨터는 'B형 + AB형 = O형이다.'라고 잘못된 관계를 맺을 수 있게 됨.
    - 별도의 column을 형성해 주고, 1개의 column에는 True, 나머지는
      모두 False라고 넣어줌으로써 'A, B, AB, O형의 관계는 독립적이다.'라는
      카테고리를 표현해 주는 방식

    - 원-핫 인코딩을 도입하는 이유임.
"""
```

```python
pd.get_dummies(df['혈액형_code'])
```
