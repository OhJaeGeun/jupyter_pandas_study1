```python
import numpy as np
import pandas as pd
```

```python
filename = 'data/korean-idol.csv'
df = pd.read_csv(filename, sep = ',')
df
```

```python
df2 = df.copy()
df2
```

```python
# 17-2. 데이터프레임 병합하기
# - concat :row나 column 기준으로 단순하게 이어 붙이기
# - merge : 특정한 고유키(unique id) 값을 기준으로 병합하기.
```

```python
a = '''
    pd.merge(left, right, on = "기준column", how = "left")
    left와 right는 변합할 두 dataframe을 대입함.
    on에는 병합 기준이 되는 column을 넣어줌.
    how에는 "left, right, inner, outer" <- 4가지 병합 방식 중 한가지를 선택함.
'''
```

```python
df_right = df2.drop([1, 3, 5, 7, 9], axis = 0)
df_right
```

```python
df_right = df_right.reset_index(drop = True)
```

```python
df_right
```

```python
df
```

```python
pd.concat([df, df_right], axis = 1)
```

```python
pd.merge(df, df_right, on = '이름', how = 'left')
```

```python
#pd.merge(df, df_right, on = '이름', how = 'right') 아래 코드와 동일
pd.merge(df_right, df, on = '이름', how = 'left')
```

```python
# inner 방식은 Dataframe에 모두 키 값이 존재하는 경우에만 병합
pd.merge(df, df_right, on = "이름", how = "inner")
```

```python
# outer 방식은 하나의 Dataframe에 모두 키 값이 존재하는 경우에만 병합
# outer 방식에서는 없는 값은 NaN으로 대입되어 짐.
pd.merge(df, df_right, on = '이름', how = 'outer')
```

```python
df_right.columns = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
df_right
```

```python
#pd.merge(df, df_right, on = '이름', how = 'outer') 불가
# df와 df_right를 병합하려면 df에는 '이름', df_right에는 'A'로 표기되어 있어서
# 기준을 left_on, right_on을 사용하여 병합할 수 있음
pd.merge(df, df_right, left_on = '이름', right_on = 'A', how = 'outer')
```
